```python
'''
I have to go back and remake the logo plot to take into account every single MauE with a CXC within the IPR region
To make this a bit easier I am going to limit this to those with a single CXC since there are about 300 ish 2-3 and a hand full
with more

I will essentially go by the species list of species with the 3 quality markers. use that to filter down the list of MauE
with 1 CXC so its only things with the 3 quality markers used in the phylogenetic anaysis 

then ill use regex to just pull the CXC and two amino acids on either side. the dataframe will be the accession number 
then the then I will create a Fasta file that is just the same information so i can work it into the rest of my code
basically the final aligned file will be

the logo plot
>asdf
XXCXCXX
>asdf
XXCXCXX
.
.
.

this will then be made into a logo plot using logomaker. If the CXC is at a terminus and therefor wont have a position before 
or after the CXC it needs to be dropped. This turned out to only be 5 MauE in the whole set.

'''
```

```python
import pandas as pd
import os
import re
```

```python
#this was a species list from the species level set that only has species with the three quality markers present
#then drops mauE entries from species not on the list
species = pd.read_excel(r"C:\PATH\5_Filtered_FBDS_IPRcounts_byspecies_MAUA_names.xlsx")
fbds = pd.read_excel(r"C:\PATH\3b_FBDS_CXC_annotated_false_pos_removed_MAUA.xlsx")
print(species.shape)
print(fbds.shape)
```

```python
#i am only looking at 1 cxc MauE
maue = fbds[fbds['CXC_at_all'] == 1].copy()
print(maue.shape)
```

```python
species_list = list(species['species'])
print(species_list[0:10])
```

```python
#in this section I am dropping anyting not in the species list I just made
#I also need to remove things that have the CXC at the N or C terminus because they wont be able to actually align 
#becuase I am including two positions before and after the CXC
species_count = 0
with open(r'C:\PATH\XXCXCXX_9925.fasta','a') as file:
    for i in maue.index:
        spec = maue.at[i,'species']
        accession = maue.at[i,'Accession']
        sequence = maue.at[i,'Sequence']
        search = re.search(r"[A-ZU][A-ZU]C[A-ZU]C[A-ZU][A-ZU]",sequence)
        if search:
            cxc = search.group()
            if spec in species_list:
                file.write('>'+accession+'\n'+cxc+'\n')
                species_count += 1
            else:
                continue
        else:
            print(accession,' has a CXC at either end')
    file.close()
print(species_count)
#delete the file as it will just append to any file with the same name if your repeating        
```
