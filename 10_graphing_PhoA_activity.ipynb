```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import starbars
```

```python
#the graph for b8flw7

b8_data = { 'K12-WT':100, 'KO-DsbB': 20.98053175,'B8FLW7-WT':164.8151277, 'B8FLW7-C1':27.68639298, 'B8FLW7-C2':33.25186839,'B8FLW7-DM':25.99778356}

b8_sd = [18.58023696,2.160138569,7.574149239,2.655810771,3.841766872,2.321951401]

mutants = list(b8_data.keys())
values = list(b8_data.values())
 
fig = plt.figure(figsize = (10, 5))

asymmetric_errors = [np.zeros(len(b8_sd)), b8_sd]
plt.errorbar(x=mutants, y=values, yerr=asymmetric_errors, fmt='none', ecolor='black', capsize=10, capthick=0.8)


plt.bar(mutants, values, color ='darkslategrey',
        width = 0.8,capsize=10, edgecolor='black', linewidth=1.5)

plt.ylabel("Relative Activity by percent")
plt.title("PhoA activity of wild-type B8FLW7 vs. single and double mutants")

annotations = [("KO-DsbB", "B8FLW7-WT", 0.0000000134897),("KO-DsbB", "B8FLW7-C2", 0.00005)]
starbars.draw_annotation(annotations)

plt.show()

fig.savefig(r'C:\PATH\Phoa_b8flw7_2p.png', dpi='figure',  bbox_inches='tight', transparent=True )
```

```python

import matplotlib.pyplot as plt
import numpy as np

species = ("K12", "KO-DsbB", "WT-MauE", 'C1','C2','DM')

#below is the data from the supplementary files
#'H1XRV2-WT':85.71742713, 'H1XRV2-C1':40.32324822, 'H1XRV2-C2':34.28675627,'H1XRV2-DM':35.02058266
#'FLW7-WT':164.8151277, 'B8FLW7-C1':27.68639298, 'B8FLW7-C2':33.25186839,'B8FLW7-DM':25.99778356
k12=100
ko=20.98053175
penguin_means2 = {
    'B8FLW7': (0, 0, 164.8151277,27.68639298,33.25186839, 25.99778356),
    'H1XRV2': (0, 0, 85.71742713,40.32324822, 34.28675627, 35.02058266) }


x = np.arange(len(species))  # the label locations
width =0.45 # the width of the bars
multiplier =0

fig, ax = plt.subplots()
fig.set_constrained_layout(True)
o2=0.24
offset = width * multiplier
err_kw={'ecolor':'black', 'capsize':10, 'capthick':0.8}

#k12
rects = ax.bar(0.46, k12, width, yerr=np.array([[0.],[18.58023696]]), error_kw=err_kw, color ='white', edgecolor='black', linewidth=1.5, label='K12')
#KOdsbb
rects = ax.bar(1.45, ko, width, yerr=np.array([[0.],[2.160138569]]),error_kw=err_kw, color ='white', edgecolor='black', linewidth=1.5, label='KO')
    
for attribute, measurement in penguin_means2.items():
    if attribute == 'B8FLW7':
                offset = width * multiplier
                asymmetric_errors = np.array([[0.,0.,0.,0.,0.,0.],[0.,0.,7.574149239,2.655810771,3.841766872,2.321951401]])
                rects = ax.bar(x + offset+o2-0.01, measurement, width, yerr=asymmetric_errors, error_kw=err_kw,  color ='grey', edgecolor='black', linewidth=1.5, label=attribute)
                multiplier += 1
    else:
                offset = width * multiplier
                asymmetric_errors = np.array([[0.,0.,0.,0.,0.,0.],[0.,0.,7.110196859,3.357949565,4.918907293,4.869949807]])
                rects = ax.bar(x + offset+o2+0.025, measurement, width, yerr=asymmetric_errors, error_kw=err_kw,  color ='black', edgecolor='black', linewidth=1.5, label=attribute)
                #ax.bar_label(rects, padding=3)
                multiplier += 1
                

ax.set_ylabel('Relative Activity by percent')
ax.set_title('PhoA activity of single and double mutants')
ax.set_xticks(x + width)
ax.legend(loc='upper right', ncol=2)
ax.set_ylim(0, 200)

plt.show()
fig.savefig(r'C:\PATH\Phoa_bargraph.png', dpi=1000,  bbox_inches='tight',  transparent=True )

```
