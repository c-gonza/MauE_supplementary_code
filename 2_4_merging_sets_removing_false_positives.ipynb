```python
#removing MauE entries that do not satisfy the CXC motif being inside the IPR region by splitting and filtering
maue_cxc_mask = (combined_set_df2['IPR009908'] == 'IPR009908') & (combined_set_df2['CXC_inside_IPR'] > 0)
true_maue = combined_set_df2.loc[maue_cxc_mask]
rest_of_set_mask = (combined_set_df2['IPR009908'] != 'IPR009908')
rest_of_set = combined_set_df2.loc[rest_of_set_mask]
display(true_maue)

display(rest_of_set)
                                                                
                    
```

```python
#merging into the filtered set
filtered_fbds = pd.concat([true_maue,rest_of_set])
```

```python
'''
The above data set is done. but species without the three quality markers are still present, the below sections give a species 
level data set that gets filtered for the 3 markers in notebook 5

'''
```

```python

#were condensing down to species counts so were dropping columns not related to phylogeny and IPS counts. at this point
#we only have true maue according to our filters
phylogeny_columns = ['superkingdom', 'clade', 'phylum', 'class', 'order', 'family', 'genus', 'strain', 'no rank', 'species group', 'subspecies', 'suborder', 'subclass', 'subgenus', 'species subgroup', 'isolate', 'tribe', 'pathogroup', 'subfamily', 'serotype', 'serogroup', 'forma specialis', 'biotype']  # Add other relevant columns here

# Group by species and count the IPR entries for that species
ipr_counts = filtered_fbds.groupby('species')[['IPR009908', 'IPR012932', 'IPR003752','IPR047873', 'IPR010243', 'IPR036286','IPR026259','IPR036560']].count()
display(ipr_counts)
# pulls the phylogeny info by taking the first Taxid for each phylo column for the grouped species. they are all the same.
phylogeny_info = combined_set_df2.groupby('species')[phylogeny_columns].first()
display(phylogeny_info)
# Combine data frames
species_summary_df = pd.concat([ipr_counts, phylogeny_info], axis=1)

# reset the index so we keep species as a column
species_summary_df.reset_index(inplace=True)

# small check
print(species_summary_df)
```
